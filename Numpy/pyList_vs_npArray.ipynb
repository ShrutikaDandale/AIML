```python
import numpy as np
import time
```

```python
# compare py_list and numpy_array on execution performance
# kitana time py_list ya numpy_arr legi execute hone me
size = 1_000_000

# python list
py_list = list(range(size))
start = time.time() #statrting time of execution
sq_list = [x**2 for x in py_list] #koi bhi operation perform kiya simple sa kyuki execution time operation ka nikalana hai
end = time.time() #ending time of execution
print(f"Python list time = {end-start} seconds")

# numpy arrays
np_arr = np.array(py_list)
start = time.time()
sq_array = np_arr ** 2 #vectorization-no need to use loop in npy
end = time.time()
print(f"Numpy array time = {end-start} seconds")
```

```python
# memory comparison of py & npy
import sys
print(f"python list size = {sys.getsizeof(py_list) * len(py_list)} bytes")
print(f"Numpy array size = {np_arr.nbytes} bytes")

# we can see difference in output
```
