```python
import numpy as np 
```

```python
# integer
arr = np.array([1, 2, 3, 4, 5])
print(arr, arr.dtype)
```

```python
# floating
arr = np.array([1, 2, 3, 4.5, 5])
print(arr, arr.dtype)
```

```python
# boolean
arr = np.array([1, 2, 3, 4, 5])
arr = arr > 3
print(arr,arr.dtype)
```

```python
# complex nums
arr1 = np.array([3 + 5j])
arr2 = np.array([4 + 6j])

print(arr1, arr1.dtype)
print(arr1 + arr2)
print(arr1 - arr2)
```

```python
# string
arr = np.array(["Apple", "Banana", "Mango"])
print(arr, arr.dtype)
```

```python
# object
arr = np.array(["prime", {1, 2, 3, 4, 5}, 3.14])
print(arr, arr.dtype)
```
