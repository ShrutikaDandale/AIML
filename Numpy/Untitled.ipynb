```python
import numpy as np
```

```python
# Aggregation functions

arr = np.array([1, 2, 3, 4, 5])

print(np.sum(arr))
print(np.prod(arr))
print(np.min(arr))
print(np.max(arr))
print(np.argmin(arr)) #prints min index in the arr
print(np.argmax(arr)) #prints max index in the arr

print(np.mean(arr))
print(np.std(arr))
print(np.median(arr))
print(np.var(arr))
```

```python
# power function
arr = np.array([1, 2, 3, 4, 5])

print(np.square(arr))
print(np.sqrt(arr))
print(np.pow(arr, 3)) #elements to the power something
```

```python
# log function

arr = np.array([1, 2, 3, 4, 5])

print(np.log(arr))
print(np.log10(arr))
print(np.log2(arr))

print(np.exp(arr))
```

```python
# rounding off function 

print(np.round(3.14)) #roun off
print(np.ceil(3.14)) #round up
print(np.floor(3.14)) #round down
print(np.trunc(3.14)) #truncate the value after decimal
```

```python
#extra functions
arr = np.array([1, -2, 3, 4, 5, -5, 2])

print(np.unique(arr)) #it will print all unique values in the arr not repeated
print(np.sort(arr))
print(np.abs(arr)) #it makes all the values positive
```
