```python
import pandas as pd
```

```python
s = pd.Series([1, 2, 3, 4, 5])
print(s)
print(type(s))

# indexing
print(s[0])
print(s[2])
```

```python
s = pd.Series([23, 24, 25, 26], index = ["Jhope", "Yoongi", "Namjoon", "Jin"])
print(s)
#here index are lable name of the values 


# acces data with lable value
print(s["Jin"])
print(s["Namjoon"])

# acces data with index value
print(s[0])
print(s[3])

print(s.index)
```
