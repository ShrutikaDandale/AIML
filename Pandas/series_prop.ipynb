```python
import pandas as pd
```

```python
# Homogenous- can strore multiple types of values in same list 
s = pd.Series([1, 2, 3, 4, "Shrutika"])
print(s)
print(type(s))

# they will have same dtype as an object
```

```python
# vectorization
s1 = pd.Series([1, 2, 3, 4])
s2 = pd.Series([1, 2, 3, 4])

print(s1 + s2)
```

```python
# mutable- change the values of the specific index
s1 = pd.Series([1, 2, 3, 4, 5])
s1[0] = 100
print(s1)
```

```python
# immutable size- we cannot change the actual size of the array
s1 = pd.Series([1, 2, 3, 4, 5])

changed_s1 = s1.drop(0) #we deop the value of index 0 here but still the total indexes are 4 only

print(s1)
print(changed_s1)

# jo value drop kiya vahi sirf hat jayega and array ka size still same hi rahega even though unke serial no.s bhi same hi rahenge
```
