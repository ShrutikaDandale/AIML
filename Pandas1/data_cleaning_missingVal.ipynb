```python
import pandas as pd
```

```python
# cleaning data
df = pd.read_csv("raw_data.csv")
df
```

```python
df.isnull() 
df.isna()# dono same hai
# jahape bhi empty value hogi vahape ye false likhake dega aur jaha value hogi vaha true
```

```python
df.isnull().sum()
# it will give the count of missing values
```

```python
# it will drop the empty value cells

df.dropna() #row drop- jis bhi row me empty value hai use drop karega
df.dropna(axis = 1) #colm drop- jis bhi colm me empty value hai use drop karega
```

```python
# it will fill the empty space with some values
df.fillna(100)

# if we want to fill the particular colm age we will take the avg of others age calculate the mean and fill the value
age_mean = df["age"].mean()
df["age"].fillna(age_mean)
```

```python
# now hume ye age apane original table ke andar fill karane hai to 
age_mean = df["age"].mean()

cleaned_data = df.copy()
age_mean - cleaned_data["age"].mean()
cleaned_data["age"] = cleaned_data["age"].fillna(age_mean)
cleaned_data
```
