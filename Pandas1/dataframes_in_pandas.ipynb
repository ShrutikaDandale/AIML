```python
import pandas as pd
```

```python
# DataFrames using the dictonary of list

info = {
    "Name" : ["Shrutika", "Devanshi", "Prachi", "Shravani"],
    "Age" : [23, 21, 24, 25],
    "CGPA" : [9.27, 8.73, 8.90, 9.07]
}

df = pd.DataFrame(info)

print(df, type(df))

print(df.index) # it shows row indices. RangeIndex(start=0, stop=4, step=1), it means index start from zero, stops means total 4 index and each are seperated using 1 step 
print(df.columns) #to access the column values
```
