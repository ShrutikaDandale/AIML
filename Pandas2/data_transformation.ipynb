```python
import pandas as pd
```

```python
df = pd.read_csv("raw_data.csv")
df
```

```python
# feature engineering
# apply() - this fxn will add the new coloumn in the existing data

df2 = df.copy()

df2["tax"] = df2["income"].apply(lambda x : "20%" if x >= 60000 else "10%")
# here it added tax colm, the tax is calculatd by using income

# map() - this fxn helps to mapped the values 
# suppose we wanna convert the Male - M, Female - F, Unknown - U we will use map() fxn
gender_map = {"Male" : "M","Female": "F", "Unknown" : "U"}
df2["gender"] = df2["gender"].map(gender_map)


# assign() - this fxn assign a new coloumn and new vals to that colm
df2 = df2.assign(new_income = df2["income"] * 1.1) 


# replce() - change and replace the value
df2["country"] = df2["country"].replace("USA", "US")


# rename
df2.columns = ["Id", "Name", "Age", "Country", "Gender", "Income", "Tax", "New_Income"]
df2.rename(columns={"Income":"Salary"})
df2.rename(index={1:"First"})


# sorting values
df2 = df2.fillna(50)
df2.sort_values("Income") #ye ascendingorder me print hoga
df2.sort_values("Income", ascending=False) #if we want in descending order
sorted_df = df2.sort_values(["Income", "Age"]) #if income dono ka same hoga to age ke basis pe sorting hogi

# index sorting
sorted_df.sort_index()

# reset() - isme humara data as it is rahega aur sirf index sort hoga
sorted_df.reset_index()
sorted_df.reset_index(drop=True) #agar hum chahte hai ki hamari index preserve na ho aur vo reset ho jaye


# ranking- assign the rank from lowest to highest
sorted_df["Ranking"] = sorted_df["Income"].rank()
sorted_df["Ranking"] = sorted_df["Income"].rank(ascending=False) #reverse in descending order highest to lowest
# it will assign the rank on the basis of the income

# sorted_df["Ranking"] = sorted_df["Income"].rank(method = "dense", method = "max" ) dense- if values are both same then assign only one, max- assign max value

sorted_df


# if we wanna switch the places of coloumns
sorted_df[["Id", "Name", "Age", "Country", "Gender", "Income", "New_Income", "Tax"]] 
# tax will get shifted to the last
```
